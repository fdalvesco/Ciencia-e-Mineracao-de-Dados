# Explique como esse gráfico:

# Por que é interessante fazer um gráfico de dispersão nesse caso?
R: Pare facilitar uma visualização mais clara da distribuição dos dados, analisar as tendências e identificar os padrões nas relações entre as variáveis.

# Qual é a diferença entre o gráfico de linha na interpretação?
R:
#####O gráfico de linha é apropriado para visualizar a evolução de uma variável ao longo do tempo.
######O gráfico de dispersão é mais útil para avaliar a relação geral entre duas variáveis, independentemente de uma dimensão temporal específica

## Modifique a distribuição aleatória para distribuição normal - Pesquise a função numpy

```python
# Vamos gerar um conjunto de dados aleatórios para as idades
np.random.seed(0)
idades = np.random.randint(0, 100, size=1000) #distribuição aleatória de valor inteiro

# Criar o histograma
plt.hist(idades, bins=20, edgecolor='black')

plt.title('Distribuição das Idades')
plt.xlabel('Idade')
plt.ylabel('Frequência')

# Mostrar o gráfico
plt.show()
```

```python
# Vamos gerar um conjunto de dados aleatórios para as idades
np.random.seed(0)
idades = np.random.normal(50, 10, size=1000) #distribuição aleatória de valor inteiro

# Criar o histograma
plt.hist(idades, bins=20, edgecolor='black')

plt.title('Distribuição das Idades')
plt.xlabel('Idade')
plt.ylabel('Frequência')

# Mostrar o gráfico
plt.show()
```

# Interprete a distribuição distribuição e compare a diferença com a distribuição normal

R: Uma distribuição aleatória é caracterizada por dados sem uma ordem específica, exibindo uma distribuição irregular em que pode haver uma grande quantidade no início e no meio do gráfico.

Por outro lado, na distribuição normal, os dados apresentam menor quantidade tanto no início quanto no final do gráfico, com uma maior concentração de valores no meio.

# Gráfico de Caixa (Box Plot)

```python
# Dados
distribuicao = np.random.normal(40, 20, 120)

# Criar o gráfico de caixa
plt.boxplot(distribuicao)

plt.title('Distribuição Personalizada')
plt.xlabel('Dados')
plt.ylabel('Valores')

# Mostrar o gráfico
plt.show()
```

# Explique o Gráfico acima -- Pesquise
R: O gráfico de caixa é uma representação gráfica que fornece informações sobre a distribuição de um conjunto de dados. É utilizado para visualizar medidas de tendência central, dispersão e a presença de valores discrepantes nos dados.

```python
!gdown 1GvcLp9qKfSI2o0CHvXjsGXDRTZ91SGy9
```

```python
df = pd.read_csv('Pontuacao.csv')
df
```

```python
# Criar o gráfico de violino
equipes = df.columns.tolist()
plt.figure(figsize=(8, 6))
plt.violinplot(df)

# Personalizar o gráfico
plt.title('Distribuição da Pontuação por Equipe')
plt.xlabel('Equipes')
plt.ylabel('Pontuação')

plt.xticks(np.arange(1, len(equipes)+1), equipes)

# Mostrar o gráfico
plt.show()
```

# Explique o gráfico acima, qual é a diferença da Equipe C da Equipe D.
R: Cada violino no gráfico representa a distribuição das pontuações de cada equipe. 
A largura do violino indica a densidade para cada equipe, sendo maior onde a frequência é maior na faixa. 
A linha branca no centro representa a mediana. 
As extremidades dos violinos representam as arestas, mostrando a variação das pontuações em cada equipe. 
Os pontos individuais acima e abaixo dos violinos representam valores discrepantes.

No gráfico, a equipe C apresenta maior discrepância de dados em comparação com a equipe D, além de exibir uma frequência diferente.
